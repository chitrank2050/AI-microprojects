# 1. Random Experiments & Sample Space

A random experiment has:
- Sample space (all possible outcomes) (i.e for dice throw sample space is [1,2,3,4,5,6])
- Events (subsets of outcomes) (i.e event of getting odd number is [1,3,5])

P(E) = n(E) / n(S)
<br/>n(E) -> Event
<br/>s(E) -> Sample space

Example:
Coin toss → {H, T}
Dice roll → {1,2,3,4,5,6}

In ML, a "random variable" represents numerical outcomes.

# 2. Random Variables

A random variable maps outcomes → numbers.

Two types:
- Discrete
- Continuous

## Discrete Random Variable

Example: Bernoulli variable (0 or 1)
<br/>
Used in classification.

```python
from scipy.stats import bernoulli

p = 0.7
rv = bernoulli(p)

print("P(X=1):", rv.pmf(1))
print("P(X=0):", rv.pmf(0))
```

## Continuous Random Variable

Example: Gaussian (Normal Distribution)

Very important in ML.
Most natural processes approximate normal distribution.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
mu = 0
sigma = 1

pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.title("Normal Distribution (μ=0, σ=1)")
plt.show()
```

# 3. Expected Value (Mean)

For discrete variable:

E[X] = Σ x P(x)

For continuous:

E[X] = ∫ x f(x) dx

Mean represents center of distribution.

```python
samples = np.random.normal(0, 1, 10000)

print("Sample mean:", np.mean(samples))
```

# 4. Variance

Variance measures spread:

Var(X) = E[(X - μ)²]

Standard deviation = sqrt(variance)

High variance → more uncertainty.

```python
print("Sample variance:", np.var(samples))
print("Sample std dev:", np.std(samples))
```

# 5. Joint Probability

Joint probability is the probability of two (or more) events happening simultaneously. It is denoted as P(A∩B) for two events A and B, which reads as the probability of both A and B occurring.

For two events A and B, the joint probability is defined as:

    P(A∩B)=P(both A and B occur)
    P(AUB)=P(both A or B occur) 

Note: If A and B are dependent, the joint probability is calculated using conditional probability

For two variables X and Y (if dependent):

P(A∩B) = P(A) * P(B|A)

If independent:

P(A∩B) = P(A)P(B)

-> p(A ∩ B) (t is the probability of the intersection of two or more events )

Independence is a strong assumption.
Naive Bayes assumes conditional independence.

# 6. Conditional Probability
Conditional probability is the probability of an event happening when another event has already occurred. It helps us adjust our prediction of an event using the information we already know.

P(A | B) = P(A ∩ B) / P(B)

This is fundamental to ML.

Where:
<br/>
    P(A∩B) is the joint probability of both events A and B occurring.
    <br/>
    P(B) is the probability of event B occurring.

Let A be the event of drawing an Ace and B be the event of drawing a red card.
<br/>
In a standard deck of 52 cards, there are 26 red cards. Out of these, 2 are red Aces (Ace of Hearts and Ace of Diamonds).
<br/>
The required probability is:
<br/>
P(A|B) = P(A ∩ B) / P(B)
<br/>
Here,
  P(A ∩ B) = 2/52 and P(B) = 26/52
<br/>
So,
  P(A|B) = (2/52) / (26/52) = 2/26 = 1/13.



```python
# simple simulation example

data = np.random.choice([0,1], size=1000, p=[0.6, 0.4])

p_A = np.mean(data == 1)
print("P(A=1):", p_A)
```

# 7. Bayes Theorem

P(A | B) = P(B | A) P(A) / P(B)

This flips conditional probabilities.

Used in:
- Naive Bayes
- Medical diagnosis
- Spam filtering

## Example: Medical Test

Suppose:
- 1% have disease
- Test accuracy = 95%

What is probability of actually having disease if test is positive?

```python
P_disease = 0.01
P_positive_given_disease = 0.95
P_positive_given_no_disease = 0.05

P_positive = (
    P_positive_given_disease * P_disease +
    P_positive_given_no_disease * (1 - P_disease)
)

P_disease_given_positive = (
    P_positive_given_disease * P_disease
) / P_positive

print("P(Disease | Positive):", P_disease_given_positive)
```

# 8. Maximum Likelihood Estimation (MLE)

In ML, we choose parameters that maximize:

P(data | parameters)

For Gaussian:

μ = sample mean
σ² = sample variance

This is how models "learn".

```python
data = np.random.normal(5, 2, 1000)

mu_hat = np.mean(data)
sigma_hat = np.std(data)

print("Estimated μ:", mu_hat)
print("Estimated σ:", sigma_hat)
```
